```python
import quandl
import os
from dotenv import load_dotenv
```

```python
# create account and get my own keys
env_file = 'C:/Users/DELL/Desktop/env_vars_dgroner_ws.env'
load_dotenv(env_file)
quandl_key = os.getenv('QUANDL_KEY')
print(quandl_key)
```

```python
# set the global quandl key variable
quandl.ApiConfig.api_key = quandl_key
```

```python
# Gets the C02 Emmissions, which has a Quandl Code of
# BP/C02_EMMISSIONS_LKA from the Energy Production and Consumption dataset: 
data = quandl.get("BP/C02_EMMISSIONS_LKA")
#print(type(data))
print(data)
```

```python
#Consolidate the data into a pandas DataFrame
import pandas as pd
df = pd.DataFrame(data,columns = ['Value'])
df
```

```python
#Visualize the data as a multi-line time series plot.
import matplotlib.pyplot as plt
plt.plot(df['Value'],alpha=0.8, label=' BP/C02_EMMISSIONS_LKA')
```
